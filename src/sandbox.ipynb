```python
# TODO exploratory data visualization

import pandas as pd 
import numpy as np 

# use MAF 01 for 22N9 
sel_n9_geno = pd.read_csv("/work/hs325/csgs2026/data/selectedlines/geno/22N9_MAF01_Genotypes_V1.csv", index_col=0)
sel_all_geno = pd.read_csv("/work/hs325/csgs2026/data/selectedlines/geno/SelectedLines_(All)_MAF01_Genotypes_V1.csv", index_col=0)
sel_n9_pheno = pd.read_csv("/work/hs325/csgs2026/data/selectedlines/pheno/22N9_Phenotypes_V1.csv")
sel_all_pheno = pd.read_csv("/work/hs325/csgs2026/data/selectedlines/pheno/SelectedLines_Phenotypes_V1.csv")

# use MAF 05 for wild
wild_22dbw_geno = pd.read_csv("/work/hs325/csgs2026/data/wildlines/geno/Wild_22DBW_MAF05_Genotypes_V1.csv", index_col=0) 
wild_all_geno = pd.read_csv("/work/hs325/csgs2026/data/wildlines/geno/Wild_Both_Corrected_Genotypes_V1.csv", index_col=0) 
wild_22dbw_pheno =pd.read_csv("/work/hs325/csgs2026/data/wildlines/pheno/Wild_(22DBW)_Phenotypes_V1.csv") 
wild_all_pheno = pd.read_csv("/work/hs325/csgs2026/data/wildlines/pheno/Wild_(Both)_Phenotypes_V1.csv") 
```

```python
sel_n9_pheno.set_index("SampleID", inplace=True)
sel_all_pheno.set_index("SampleID", inplace=True)
wild_22dbw_pheno.set_index("SampleID", inplace=True)
wild_all_pheno.set_index("SampleID", inplace=True)
```

```python
# check shapes
print("sel_n9_geno shape:", sel_n9_geno.shape)
print("sel_all_geno shape:", sel_all_geno.shape)
print("sel_n9_pheno shape:", sel_n9_pheno.shape)
print("sel_all_pheno shape:", sel_all_pheno.shape)
print("wild_22dbw_geno shape:", wild_22dbw_geno.shape)
print("wild_all_geno shape:", wild_all_geno.shape)
print("wild_22dbw_pheno shape:", wild_22dbw_pheno.shape)
print("wild_all_pheno shape:", wild_all_pheno.shape)
```

```python
pairs = [
    ("Selected N9", sel_n9_geno, sel_n9_pheno),
    ("Selected All", sel_all_geno, sel_all_pheno),
    ("Wild 22DBW", wild_22dbw_geno, wild_22dbw_pheno),
    ("Wild All", wild_all_geno, wild_all_pheno)
]

for name, geno, pheno in pairs:
    # Use the first column if geno index is still numeric, otherwise use geno.index
    geno_ids = set(geno.iloc[:, 0]) if isinstance(geno.index, pd.RangeIndex) else set(geno.index)
    pheno_ids = set(pheno.index)
    
    # Bidirectional difference
    missing_in_pheno = geno_ids - pheno_ids
    missing_in_geno = pheno_ids - geno_ids
    
    print(f"=== {name} ===")
    
    # 1. Genotype IDs missing in Phenotype
    print(f"Number of genotype IDs missing in phenotype: {len(missing_in_pheno)}")
    if len(missing_in_pheno) > 0:
        print("Missing in Phenotype:", sorted(list(missing_in_pheno)))
        
    # 2. Phenotype IDs missing in Genotype
    print(f"Number of phenotype IDs missing in genotype: {len(missing_in_geno)}")
    if len(missing_in_geno) > 0:
        print("Missing in Genotype:", sorted(list(missing_in_geno)))
        
    print("\n" + "="*30 + "\n")
```
